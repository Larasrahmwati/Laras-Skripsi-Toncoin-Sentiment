```python
# ===============================
# INSTALL LIBRARY
# ===============================
!pip install langdetect deep-translator nltk Sastrawi -q

import re
import pandas as pd
import nltk
from langdetect import detect
from deep_translator import GoogleTranslator
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from Sastrawi.Stemmer.StemmerFactory import StemmerFactory

# Download resource NLTK
nltk.download('punkt')
nltk.download('punkt_tab')
nltk.download('stopwords')

if 'data_clean' not in locals():
    data_clean = data.drop_duplicates(
        subset=['created_at', 'full_text', 'favorite_count', 'retweet_count']
    ).copy()

# ===============================
# TRANSLASI KE BAHASA INDONESIA
# ===============================
def translate_to_indonesian(text):
    try:
        if detect(str(text)) != 'id':
            return GoogleTranslator(source='auto', target='id').translate(text)
        return text
    except:
        return text

data_clean['text_translated'] = data_clean['full_text'].apply(translate_to_indonesian)

# ===============================
# PREPROCESSING
# ===============================

# Inisialisasi Stopwords
stop_words = set(stopwords.words('indonesian')) | set(stopwords.words('english'))
stopwords_lokal = {
    'di', 'yang', 'dan', 'dari', 'ke', 'ini', 'itu', 'untuk',
    'pada', 'adalah', 'dengan', 'saya', 't', 'co', 'https', 'http',
    'rt', 'rw', 'nan', 'terkait', 'bisa', 'akan', 'sudah'
}
stop_words.update(stopwords_lokal)

# Stemmer Indonesia
factory = StemmerFactory()
stemmer = factory.create_stemmer()

def preprocess_text(text):
    # Case folding
    text = str(text).lower()

    # Cleaning (hapus URL, mention, hashtag, angka, simbol)
    text = re.sub(r"http\S+|www\S+|@\w+|#\w+", "", text)
    text = re.sub(r"[^a-z\s]", "", text)

    # Tokenizing
    tokens = word_tokenize(text)

    # Stopword removal
    tokens = [word for word in tokens if word not in stop_words]

    # Stemming
    tokens = [stemmer.stem(word) for word in tokens]

    return " ".join(tokens)

# Terapkan preprocessing
data_clean['clean_text'] = data_clean['text_translated'].apply(preprocess_text)

# ===============================
# TAMPILKAN HASIL
# ===============================
display(
    data_clean[['full_text', 'text_translated', 'clean_text']].head()
)

print("Jumlah total data:", data_clean.shape[0])
```

```python
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer

# 1. Initialize TfidfVectorizer
vectorizer = TfidfVectorizer()

# 2. Fit and transform the cleaned text to create the tfidf_matrix
# Ensure data_clean['text_clean'] exists from previous preprocessing steps
tfidf_matrix = vectorizer.fit_transform(data_clean['text_translated'])

# 3. Define X (features) and y (target) for splitting
X = tfidf_matrix
y = data_clean['sentiment']

# 4. Split the X and y data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.2,    # 20% data for testing
    random_state=42,  # For reproducibility
    stratify=y        # To maintain sentiment distribution
)

print("Data Training (80%):")
print(y_train.value_counts())

print("\nData Testing (20%):")
print(y_test.value_counts())
```
