```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('TON_1TAHUN - Data_TonCoin1 (1).csv')

df.head()
```

```python
print("Jumlah baris, kolom:", df.shape)
print(df.columns.tolist())
```

```python
df.columns = df.columns.str.replace(';', '', regex=False)
df.columns = df.columns.str.strip()

print(df.columns.tolist())
```

```python
df = df.replace(';+', '', regex=True)
```

```python
df = df.replace(r'^\s*$', np.nan, regex=True)
```

```python
df = df[df['full_text'].astype(str).str.strip() != '']
```

```python
df = df.drop_duplicates(subset=['full_text'])
df = df.reset_index(drop=True)
```

```python
print("Ukuran data setelah cleaning:", df.shape)
df.head()
```

```python
from deep_translator import GoogleTranslator
from tqdm import tqdm

translator = GoogleTranslator(source='auto', target='id')

translated = []

for text in tqdm(df['full_text']):
    try:
        translated.append(translator.translate(str(text)))
    except:
        translated.append(text)

df['full_text_indonesia'] = translated
```

```python
df.to_csv('hasil_clean_translate.csv', index=False, encoding='utf-8-sig')
```
